```python
import imageio
import math
import numpy as np # numpy é uma biblioteca de operações matemáticas
from PIL import Image, ImageFilter, ImageOps # PIL - Python Image Library
from skimage import util

# Carrega uma imagem para a variavel 'im'. No caso de multiplas imagens, se utiliza um FOR.
im = Image.open("/Users/arielecamara/Desktop/Mestrado/grandedolmen.jpg") # im é o nome da váriavel da minha imagem
```

```python
# Próximo passo refazer para mais imagens
# Adicionar linhas no DataFrame. O Datframe tem várias linhas, e uma linha é para cada imagem.
# O DataIku será utilizado para o treino do DATASET e para a classificação das imagens. 
```

```python
import numpy as np 
from PIL import Image 
import os
import pandas as pd 
dataset = []
for filename in os.listdir("/Users/arielecamara/Desktop/Mestrado/Imagens-Dólmens"):
    im = Image.open("/Users/arielecamara/Desktop/Mestrado/Imagens-Dólmens/" + filename)

    im = im.convert("1") #imagem em preto OU branco (1 e 0)
    size = 256,256 
    im = im.resize(size, Image.ANTIALIAS) 
    im_array = list ([list(x)for x in np.asarray(im).astype(int)]) 
    
    lista2 = []
    for i in im_array: 
        lista2.append("".join(list(map(lambda x : "".join(str(x)),i)))) 
    lista2 = list(map(lambda x : int(x,2),lista2))
    dataset.append(lista2)

# Passar para uma variavel pandas - criar primeiro o nome das colunas que vou atribuir
labels = list(range(0,256))
labels = list(map(lambda x : str(x), labels))

# Criar lista em panda
df = pd.DataFrame(np.array(dataset).reshape(len(dataset),256), columns = labels)
print(df)
    
```
